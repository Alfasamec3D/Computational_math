### Часть 1. Модель хищник-жертва

Рассмотрим уравнения Лотки — Вольтерры для динамики популяции хищников и животных-жертв (классический пример системы ОДУ). Уравнения следующие: $x'(t) = ax - bxy$ и $y'(t) = cxy - dy$, где $x(t)$ - количество животных - жертв, а $y(t)$ - количество животных-хищников. Коэффициенты a, b, c и d описывают скорость процессов в модели. Например, $a$ - это скорость, с которой рождаются хищные животные, а $d$ - скорость, с которой хищные животные умирают. Коэффициенты $b$ и $c$ - это скорость, с которой хищники поедают добычу, и скорость, с которой популяция хищников растет за счет популяции жертвы, соответственно. Обратите внимание, что это нелинейная система ОДУ из-за членов $xy$.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# параметры модели
a, b, c, d = 0.4, 0.002, 0.001, 0.7
# правая часть системы
def f(xy_vec, t):
    x, y = xy_vec
    return [a*x - b*x*y, c*x*y - d*y]

# начальные условия
xy_vec0 = [600, 400]
# сетка
t = np.linspace(0, 50, 101)
# решение пакетным методом
xy_t = integrate.odeint(f, xy_vec0, t)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].plot(t, xy_t[:,0], 'r', label="Жертвы")
axes[0].plot(t, xy_t[:,1], 'b', label="Хищники")
axes[0].set_xlabel("Время")
axes[0].set_ylabel("Количество животных")
axes[0].legend()
axes[1].plot(xy_t[:,0], xy_t[:,1], 'ko')
axes[1].set_xlabel("Количество жертв")
axes[1].set_ylabel("Количество хищников")
```

### Часть 2. Метод высокого порядка

Задание: 
1. выписать расчетные формулы для задачи из Части 1 и реализовать метод РК первого и четвертого порядка (не в общем для любой задачи Коши, а для конкретной из Части 1)
2. решить задачу этими методами в изначальной постановке. Сравнить решение с решением с помощью odeint. Оценить влияние точности метода на решение

***ваш ответ***

$$X=
\begin{pmatrix}
x \\ y
\end{pmatrix}$$
$$F=
\begin{pmatrix}
ax-bxy \\ cxy-dy
\end{pmatrix}$$

Метод Рунге-Кутты 1-го порядка
Для шага h:
$$X_{n+1}=X_n+hF(X_n)$$


Метод Рунге Кутты 4-го порядка

$$k_1=F(X_n)$$
$$k_2=F(X_n+\frac{h}{2}k_1)$$
$$k_3=F(X_n+\frac{h}{2}k_2)$$
$$k_4=F(X_n+h k_3)$$

$$X_{n+1}=X_n+\frac{h}{6}(k_1+2k_2+2k_3+k_4)$$


Анализ

Метод RK1 (Эйлер) даёт значительную фазовую ошибку: траектория «раздувается», период и амплитуда искажаются. Требует очень маленького шага.

RK4 хорошо совпадает с odeint даже при шаге h = 0.05. Форма замкнутых орбит сохраняется, ошибка маленькая.

odeint использует адаптивный шаг и контроль локальной ошибки, поэтому его можно считать эталоном.

```python
# ваш код

import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# параметры модели
a, b, c, d = 0.4, 0.002, 0.001, 0.7

def F(X):
    x, y = X
    return np.array([a*x - b*x*y, c*x*y - d*y])



# Метод Рунге–Кутты 1-го порядка

def rk1_step(X, h):
    return X + h * F(X)

def rk1_solve(X0, t):
    h = t[1] - t[0]
    X = np.zeros((len(t), 2))
    X[0] = X0
    for n in range(len(t)-1):
        X[n+1] = rk1_step(X[n], h)
    return X



# Метод Рунге–Кутты 4-го порядка

def rk4_step(X, h):
    k1 = F(X)
    k2 = F(X + 0.5*h*k1)
    k3 = F(X + 0.5*h*k2)
    k4 = F(X + h*k3)
    return X + h*(k1 + 2*k2 + 2*k3 + k4)/6

def rk4_solve(X0, t):
    h = t[1] - t[0]
    X = np.zeros((len(t), 2))
    X[0] = X0
    for n in range(len(t)-1):
        X[n+1] = rk4_step(X[n], h)
    return X



# Решение на сетке

X0 = np.array([600, 400])
t = np.linspace(0, 50, 1001)  # довольно тонкая сетка

# решение нашими методами
rk1_sol = rk1_solve(X0, t)
rk4_sol = rk4_solve(X0, t)

# решение odeint (эталон)
ode_sol = integrate.odeint(lambda X, t: F(X), X0, t)


# Построение графиков

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.plot(t, ode_sol[:,0], 'k', label="odeint")
plt.plot(t, rk4_sol[:,0], 'r--', label="RK4")
plt.plot(t, rk1_sol[:,0], 'b:', label="RK1")
plt.title("Жертвы x(t)")
plt.legend()

plt.subplot(1,3,2)
plt.plot(t, ode_sol[:,1], 'k', label="odeint")
plt.plot(t, rk4_sol[:,1], 'r--', label="RK4")
plt.plot(t, rk1_sol[:,1], 'b:', label="RK1")
plt.title("Хищники y(t)")
plt.legend()

plt.subplot(1,3,3)
plt.plot(ode_sol[:,0], ode_sol[:,1], 'k', label="odeint")
plt.plot(rk4_sol[:,0], rk4_sol[:,1], 'r--', label="RK4")
plt.plot(rk1_sol[:,0], rk1_sol[:,1], 'b:', label="RK1")
plt.title("Фазовый портрет")
plt.legend()

plt.tight_layout()
plt.show()


# Оценка ошибок (нормы разности)

err_rk1 = np.linalg.norm(rk1_sol - ode_sol) / np.linalg.norm(ode_sol)
err_rk4 = np.linalg.norm(rk4_sol - ode_sol) / np.linalg.norm(ode_sol)

print("Относительная ошибка RK1:", err_rk1)
print("Относительная ошибка RK4:", err_rk4)
```

### Часть 3. Порядок точности метода

Найти численно и построить графики порядка методов из предыдущей части в зависимости от шага сетки в широком диапазоне шага. Объяснить, что мы видим на рисунке. Исследование порядка можно проводить на модельной задаче (не из постановки в задании).

```python
# ваш код


import numpy as np
import matplotlib.pyplot as plt

# модельная задача (угасание) 
# y' = -lambda * y, y(0)=1  => y(t)=exp(-lambda t)
lam = 5.0
y0 = 1.0
t0, tf = 0.0, 1.0

def f(y, t):
    return -lam * y

def y_exact(t):
    return np.exp(-lam * t)

# реализации шагов RK1 и RK4 для скалярной задачи
def rk1_step(y, h):
    return y + h * f(y, None)

def rk4_step(y, h):
    k1 = f(y, None)
    k2 = f(y + 0.5*h*k1, None)
    k3 = f(y + 0.5*h*k2, None)
    k4 = f(y + h*k3, None)
    return y + h*(k1 + 2*k2 + 2*k3 + k4)/6.0

def integrate_method(step_func, y0, h, t0, tf):
    N = int(np.round((tf - t0)/h))
    y = y0
    t = t0
    for i in range(N):
        y = step_func(y, h)
        t += h
    # если бы (tf - t) != 0 мы могли бы взять один последний меньший шаг, но мы выбрали h делящий интервал
    return y

# исследуем ошибки на сетке шагов
# набор шагов: h = 2^{-k}, k from 1..12 (широкий диапазон)
ks = np.arange(1, 13)  # 2^-1 ... 2^-12
hs = 2.0**(-ks)

errors_rk1 = []
errors_rk4 = []

for h in hs:
    y_rk1 = integrate_method(rk1_step, y0, h, t0, tf)
    y_rk4 = integrate_method(rk4_step, y0, h, t0, tf)
    y_ex = y_exact(tf)
    errors_rk1.append(abs(y_rk1 - y_ex))
    errors_rk4.append(abs(y_rk4 - y_ex))

errors_rk1 = np.array(errors_rk1)
errors_rk4 = np.array(errors_rk4)

# оценка эмпирического порядка между соседними шагами
# p_i = log(E_i / E_{i+1}) / log(h_i / h_{i+1})
def empirical_order(h_arr, err_arr):
    ps = []
    h_mid = []
    for i in range(len(h_arr)-1):
        if err_arr[i+1] == 0 or err_arr[i] == 0:
            ps.append(np.nan)
        else:
            p = np.log(err_arr[i] / err_arr[i+1]) / np.log(h_arr[i] / h_arr[i+1])
            ps.append(p)
        h_mid.append((h_arr[i] + h_arr[i+1]) / 2.0)
    return np.array(h_mid), np.array(ps)

h_mid_rk1, p_rk1 = empirical_order(hs, errors_rk1)
h_mid_rk4, p_rk4 = empirical_order(hs, errors_rk4)


print("\nЭмпирический порядок (между соседними шагами):")
print(" h_mid\t\t p_RK1\t\t p_RK4")
for hm, pr1, pr4 in zip(h_mid_rk1, p_rk1, p_rk4):
    print(f"{hm:.1e}\t {pr1: .3f}\t {pr4: .3f}")


plt.semilogx(h_mid_rk1, p_rk1, 'o-', label='RK1 p(h)')
plt.semilogx(h_mid_rk4, p_rk4, 's-', label='RK4 p(h)')
plt.axhline(1, color='gray', lw=0.8, ls='--')
plt.axhline(4, color='gray', lw=0.8, ls='--')
plt.xlabel('шаг h (среднее между соседними)')
plt.ylabel('наблюдаемый порядок p')
plt.title('Эмпирический порядок сходимости')
plt.legend()
plt.grid(True, which='both', ls=':')

plt.tight_layout()
plt.show()
```

***ваш ответ***

В зоне, где действие теорем о сходимости выполнено (асимптотическая зона), числа p(h) близки к теоретическим: ~1 для RK1, ~4 для RK4.

Для больших h (грубая сетка) наблюдаемые p «плывут» и могут быть меньше ожидаемого или даже отрицательными (если метод нестабилен на этих шагах).
